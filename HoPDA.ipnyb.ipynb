```python
import pandas as pd
import numpy as np

fTL = "Resources/purchase_data.csv" #"file To Load"

purchaseData = pd.read_csv(fTL)

df_purchaseData = pd.DataFrame(purchaseData) #"purchaseData" is now a Data Frame under "df_purchaseData"
df_purchaseData.head()
```

```python
df_playerList = df_purchaseData["SN"]
df_playerCount = df_playerList.unique() 

#Create List of Dictionaries, convert to df_
playerTotal = [{"Total Players": len(df_playerCount)}]
df_playerTotal = pd.DataFrame(playerTotal)
df_playerTotal
```

```python
df_itemList = df_purchaseData["Item ID"]
df_itemCount = df_itemList.value_counts()
uniqueItemCount = len(df_itemCount)
numPurchases = len(df_itemList)


df_revinueList = df_purchaseData["Price"]
revinueAverage = round(df_revinueList.mean(), 2)
revinueTotal = round(df_revinueList.sum(), 2)


purchaseAnalysis = [{"Number of Unique Items":uniqueItemCount, "Average Price":revinueAverage, 
                     "Number of Purchases":numPurchases, "Total Revenue":revinueTotal}]

df_purchaseAnalysis = pd.DataFrame(purchaseAnalysis)
df_purchaseAnalysis
```

```python
df_uniqueUserList = df_purchaseData.drop_duplicates(subset="SN", keep='first', inplace=False)

df_genderList = df_uniqueUserList["Gender"]
genderCount = df_genderList.value_counts()
playerPerc = (genderCount/ len(df_playerCount))*100


df_genderCount = pd.DataFrame(genderCount)
df_genderCount["Percentage of Players"] = playerPerc


df_genderCountsFinal = df_genderCount.rename(columns={"Gender": "Total Count", "PPer": "Percentage of Players"})
df_genderCountsFinal
```

```python
df_allmaleList = df_purchaseData.loc[df_purchaseData["Gender"] == "Male", :]
df_uqmaleList = df_allmaleList.drop_duplicates(subset="SN", keep='first', inplace=False)

df_allfemaleList = df_purchaseData.loc[df_purchaseData["Gender"] == "Female", :]
df_uqfemaleList = df_allfemaleList.drop_duplicates(subset="SN", keep='first', inplace=False)

df_allotherList = df_purchaseData.loc[df_purchaseData["Gender"] == "Other / Non-Disclosed", :]
df_uqotherList = df_allotherList.drop_duplicates(subset="SN", keep='first', inplace=False)

df_allgenderList = df_purchaseData["Gender"]
allgenderCount = df_allgenderList.value_counts()


df_maleitemList = df_allmaleList["Item ID"]
df_maleitemCount = df_maleitemList.value_counts()
uniqueMaleCount = len(df_maleitemCount)
numMalePurchases = len(df_uqmaleList)
df_malerevinueList = df_allmaleList["Price"]
malerevinueAverage = round(df_malerevinueList.mean(), 2)
malerevinueTotal = round(df_malerevinueList.sum(), 2)
purchasePerMale = round((malerevinueTotal/numMalePurchases),2)

malepurchaseAnalysis = [{"Gender": "Male", "Purchase Count": numMalePurchases, 
                         "Average Purchase Price":malerevinueAverage, "Total Purchase Value":malerevinueTotal, 
                         "Average Purchase Per Person": purchasePerMale}]

df_malepurchaseAnalysis = pd.DataFrame(malepurchaseAnalysis)


df_femaleitemList = df_allfemaleList["Item ID"]
df_femaleitemCount = df_femaleitemList.value_counts()
uniqueFemaleCount = len(df_femaleitemCount)
numFemalePurchases = len(df_uqfemaleList)
df_femalerevinueList = df_allfemaleList["Price"]
femalerevinueAverage = round(df_femalerevinueList.mean(), 2)
femalerevinueTotal = round(df_femalerevinueList.sum(), 2)
purchasePerFemale = round((femalerevinueTotal/numFemalePurchases),2)

femalepurchaseAnalysis = [{"Gender": "Female", "Purchase Count": numFemalePurchases, 
                           "Average Purchase Price":femalerevinueAverage, "Total Purchase Value":femalerevinueTotal, 
                           "Average Purchase Per Person": purchasePerFemale}]

df_femalepurchaseAnalysis = pd.DataFrame(femalepurchaseAnalysis)


df_otheritemList = df_allotherList["Item ID"]
df_otheritemCount = df_otheritemList.value_counts()
uniqueotherCount = len(df_otheritemCount)
numotherPurchases = len(df_uqotherList)
df_otherrevinueList = df_allotherList["Price"]
otherrevinueAverage = round(df_otherrevinueList.mean(), 2)
otherrevinueTotal = round(df_otherrevinueList.sum(), 2)
purchasePerother = round((otherrevinueTotal/numotherPurchases),2)

otherpurchaseAnalysis = [{"Gender": "Other/DNE", "Purchase Count": numotherPurchases, 
                          "Average Purchase Price":otherrevinueAverage, "Total Purchase Value":otherrevinueTotal, 
                          "Average Purchase Per Person": purchasePerother}]

df_otherpurchaseAnalysis = pd.DataFrame(otherpurchaseAnalysis)


conc_analysis = malepurchaseAnalysis + femalepurchaseAnalysis + otherpurchaseAnalysis
df_conc_analysis =pd.DataFrame(conc_analysis) 
df_conc_analysis_final = df_conc_analysis[["Gender", "Purchase Count", "Average Purchase Price", 
                                             "Total Purchase Value", "Average Purchase Per Person"]]
df_conc_analysis_final = df_conc_analysis_final.set_index("Gender")
df_conc_analysis_final
```

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-40", "40+",]


df_AllAgeGroups = df_purchaseData.drop_duplicates(subset="SN", keep='first', inplace=False)
df_AllAgeGroups_cut = pd.cut(df_AllAgeGroups["Age"], bins, labels=group_names)
df_AllAgeGroups["Age Group"] = df_AllAgeGroups_cut

df_binnedPurchases = df_purchaseData
df_binnedPurchases_cut = pd.cut(df_binnedPurchases["Age"], bins, labels=group_names)
df_binnedPurchases["Age Group"] = df_binnedPurchases_cut


df_binnedPurchases_grouped = df_binnedPurchases.groupby(by='Age Group')
ageSummary = df_binnedPurchases_grouped.size()
df_ageSummary = pd.DataFrame(ageSummary)

df_AllAgeGroups_grouped = df_AllAgeGroups.groupby(['Age Group'])
uQageSummary = df_AllAgeGroups_grouped.size()
df_uQageSummary = pd.DataFrame(uQageSummary)

agePercent = round(((df_uQageSummary/len(df_playerCount))*100),2)
df_agePercent = pd.DataFrame(agePercent)


df_fullSummary = df_uQageSummary
df_fullSummary["Percentage of Players"] = df_agePercent
df_fullSummary.columns = ['Total Players', 'Percentage of Players']

df_fullSummary
```

```python
df_binnedrevinueList = df_binnedPurchases_grouped["Price"] 

numbinnedPurchases = df_binnedPurchases_grouped.size()
df_numbinnedPurchases = pd.DataFrame(numbinnedPurchases)

binnedrevinueAverage = round(df_binnedrevinueList.mean(), 2) 
df_binnedrevinueAverage = pd.DataFrame(binnedrevinueAverage)

binnedrevinueTotal = round(df_binnedrevinueList.sum(), 2) 
df_binnedrevinueTotal = pd.DataFrame(binnedrevinueTotal)

purchasePerBin = round((binnedrevinueTotal/numbinnedPurchases),2) 
df_purchasePerBin = pd.DataFrame(purchasePerBin)


df_FinalBinAnalysis = df_numbinnedPurchases
df_FinalBinAnalysis["Average Purchase Price"] = df_binnedrevinueAverage
df_FinalBinAnalysis["Total Purchase Price"] = df_binnedrevinueTotal
df_FinalBinAnalysis["Average Total Purchase Per Person"] = df_purchasePerBin
df_FinalBinAnalysis.columns = ["Purchase Count", 'Average Purchase Price', 'Total Purchase Price', 'Average Total Purchase Per Person']
df_FinalBinAnalysis
```

```python
df_SNsort = df_purchaseData.sort_values('SN')
df_SNsortUq = df_SNsort.drop_duplicates(subset="SN", keep='first', inplace=False)


theSN = []
theSum = []
thePurchaseCount = []

for SNs in df_SNsortUq["SN"]:
    
    df_Player = df_purchaseData.loc[df_purchaseData['SN'] == SNs, :]
    purchaseCount = len(df_Player["Purchase ID"].value_counts())
    price = df_Player["Price"]
    theSN.append(SNs)
    theSum.append(price.sum())
    thePurchaseCount.append(purchaseCount)
    

sr_theSN = pd.Series(theSN)
sr_theSum = pd.Series(theSum)
sr_thePurchaseCount = pd.Series(thePurchaseCount)


df_totalList = pd.DataFrame(sr_theSN)
df_totalList["Purchase Count"] = thePurchaseCount

theAverageList = round((sr_theSum / sr_thePurchaseCount), 2)
df_totalList["Average Purchase Price"] = theAverageList

df_totalList["Total Purchase Value"] = theSum


df_totalList.columns = ["SN", "Purchase Count", "Average Purchase Price", "Total Purchase Value"]
df_totalList = df_totalList.set_index("SN")
df_totalList_Psorted = df_totalList.sort_values("Total Purchase Value", ascending=False)

df_totalList_Psorted.head()
```

```python
df_itemGroups = df_purchaseData.groupby(by='Item ID')
df_itemGroups_size = df_itemGroups.size()
df_totalItemList = pd.DataFrame(df_itemGroups_size)
df_itemPrices = df_itemGroups["Price"].sum()


df_IDsortUq = df_purchaseData.drop_duplicates(subset="Item ID", keep='first', inplace=False)
IDsortUq_JustIDs = df_IDsortUq["Item ID"]

df_itemsByID = pd.DataFrame(IDsortUq_JustIDs)

df_itemsByID["Item Name"] = df_IDsortUq["Item Name"]


df_itemsByID = df_itemsByID.set_index("Item ID")


df_itemsByID["Purchase Count"] = df_totalItemList

df_itemsByID["Total Purchase Value"] = df_itemPrices

df_itemsByID = df_itemsByID.sort_values('Purchase Count', ascending=False)
df_itemsByID.head()
```
